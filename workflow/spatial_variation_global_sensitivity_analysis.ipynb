## Decompose the time series

```python
# some variables for surrogate
var_explained = len(variables) #for PCA

# randomization
random_seeds = range(111,999+111,111)
```

```python
pca_data = {}
for var in model_data:
    trainY = model_data[var]['train']
    testY = model_data[var]['test']
    # Decompose the spatiotemporal series using PCA/KL
    pca_obj = PCA(n_components=var_explained, 
                  random_state=random_seeds[0], 
                  whiten=True)
    pca_obj.fit(trainY)

    print(pca_obj.explained_variance_ratio_)
    print(pca_obj.explained_variance_ratio_.sum())
    print(f'number of eigenmodes is {pca_obj.n_components_}')
    
    pca_data[var] = {}
    # get the outputs from the PCA
    pca_data[var]['klxi_train'] = pca_obj.transform(trainY)
    pca_data[var]['klxi_test']  = pca_obj.transform(testY)
    pca_data[var]['eigenratio'] = pca_obj.explained_variance_ratio_
    pca_data[var]['neig'] = pca_obj.n_components_
    pca_data[var]['eigenvalues'] = pca_obj.explained_variance_
    pca_data[var]['eigenmodes'] = pca_obj.components_
    pca_data[var]['mean_vector'] = pca_obj.mean_
```

```python
# function to upscale the WRF domain to reduce computational load
def upscale(coarsness,input_array_list,array_3d=None):
    # determine input shape and new shape for coarsening
    shape = array(input_array_list[0].shape, dtype=float)
    new_shape = coarseness * ceil(shape / coarseness).astype(int)

    output_array_list = []
    for array2d in input_array_list:

        # Create the zero-padded array and assign it with the old values
        zp_temp = zeros(new_shape)
        zp_temp[:int(shape[0]), :int(shape[1])] = array2d
        zp_temp[int(shape[0])::,:int(shape[1])] = array2d[-1,:].reshape(1,-1)
        zp_temp[:int(shape[0]),int(shape[1])::] = array2d[:,-1].reshape(-1,1)
        zp_temp[int(shape[0])::,int(shape[1])::] = array2d[-1,-1]

        temp = zp_temp.reshape((new_shape[0] // coarseness, coarseness,
                                new_shape[1] // coarseness, coarseness))
        
        output_array_list.append(temp.mean(axis=(1,3)))
    
    if array_3d is None:
        return output_array_list
    
    # this array has an extra dimension along the 0 axis.
    zp_temp = zeros((array_3d.shape[0],new_shape[0],new_shape[1]))
    zp_temp[:,:int(shape[0]), :int(shape[1])] = array_3d
    zp_temp[:,int(shape[0])::,:int(shape[1])] = array_3d[:,-1,:].reshape(array_3d.shape[0],1,-1)
    zp_temp[:,:int(shape[0]),int(shape[1])::] = array_3d[:,:,-1].reshape(array_3d.shape[0],-1,1)
    for ei in range(array_3d.shape[0]):
        zp_temp[ei,int(shape[0])::,int(shape[1])::] = array_3d[ei,-1,-1]

    temp = zp_temp.reshape((-1,
                        new_shape[0] // coarseness, coarseness,
                        new_shape[1] // coarseness, coarseness))

    array_3d_out = temp.mean(axis=(2,4))

    return output_array_list, array_3d_out
```

```python
var = 'T2m'
# coarsen the WRF grid to make more computationally efficient

# reshape into gridded form and average along time dimension
mean_vector_reshape2_timeav = pca_data[var]['mean_vector'].reshape(-1,wrflon.shape[0],wrflon.shape[1]).mean(axis=0)
eigenmodes_reshape2_timeav = pca_data[var]['eigenmodes'].reshape(-1,ntime2[var],wrflon.shape[0],wrflon.shape[1]).mean(axis=1)
# print out original shape
print(eigenmodes_reshape2_timeav.shape)
print(mean_vector_reshape2_timeav.shape)

#coarsen the grid
coarseness = 3
output_2d, eigenmodes_reshape2_timeav_upscaled = upscale(
    coarseness,[mean_vector_reshape2_timeav, wrflon, wrflat],array_3d=eigenmodes_reshape2_timeav,
)
mean_vector_reshape2_timeav_upscaled = output_2d[0]
wrflon_upscaled = output_2d[1]
wrflat_upscaled = output_2d[2]
# check new shape  
print(mean_vector_reshape2_timeav_upscaled.shape)
print(eigenmodes_reshape2_timeav_upscaled.shape)

# kl_dict for coarsened grid
kl_dict[var] = {'eigenmodes': eigenmodes_reshape2_timeav_upscaled.reshape(pca_data[var]['neig'],-1),
           'eigenvalues': pca_data[var]['eigenvalues'],
           'mean_vector': mean_vector_reshape2_timeav_upscaled.flatten(),
          }

# just dummy for the grid coarsening method
point_array[var] = range(0,eigenmodes_reshape.shape[2],1)
#
# use WRF mask to make kl_dict for lake and land..
#kl_dict[var] = {'eigenmodes': eigenmodes_reshape[:,:,point_array[var]].mean(axis=1),
#           'eigenvalues': pca_data[var]['eigenvalues'],
#           'mean_vector': mean_vector_reshape[:,point_array[var]].mean(axis=0), 
#          }
```

### GSA for LST and T2m in space

```python
# assign and execute the sensitivity computation in parallel
def add_helper(queue, pn, a1, a2, a3, a4): # the func called in child processes
    sens_sobol, srange = compute_sensitivities(a1, a2, sample_size=a3, kl_dict=a4)
    queue.put([sens_sobol, srange, pn])

q = Queue()
processes = []
proc_num = 0
sample_size = 2**8
for var in ['LST','T2m']:
    for sur in kl_surrogate[var]['NN']:
        p = Process(target=add_helper, args=(q, proc_num, sur, x_vals['train'], sample_size, kl_dict[var]))
        processes.append(p)
        p.start()
        proc_num += 1
        
rets = [None] * proc_num
for p in processes:
    ret = q.get() 
    print(f'proc_num = {ret[-1]}')
    rets[ret[-1]] = ret[0:2] # add in the right order based on proc_number
for p in processes:
    p.join()
```

```python
# collect all uncertainty and sensitivity data
SENSD_spatial = {}
nn = 0
for var in ['LST','T2m']: # loop over variables 
    print(var)
    sens_info = {}
    tvs = {} #'main': [], 'total': []}
    for ss, sur in enumerate(kl_surrogate[var]['NN']):
        sens_sobol, srange = rets[nn]
        if ss == 0:
            gsa_range = srange
        else:
            gsa_range = append(gsa_range,srange,axis=0)

        for sens_label in ['main','total']:
            if ss == 0:
                tvs[sens_label] = sens_sobol[sens_label]
            else:
                tvs[sens_label] += sens_sobol[sens_label]
        nn += 1
        
    for sens_label in ['main','total']:           
        tvs[sens_label] /= len(kl_surrogate[var]['NN'])

    SENSD_spatial[var] = {'gsa_range': gsa_range, 'tvs': tvs}
```

## Comparison to climate projections

### Read in projections, upscale and append

```python
projection_dir = '/lus/eagle/projects/COMPASS-GLM/CMIP6_PGW_Deltas/midcentury/'
projection_files = {}
projection_files['T2m'] = glob(projection_dir + '*_tas.nc')
projection_files['LST'] = glob(projection_dir + '*_ts.nc')
```

```python
delta_ens = {}
var = 'T2m'
for ff, proj_fn in enumerate(projection_files[var]):
    proj_ds = open_dataset(proj_fn)
    delta_time_mean = proj_ds.tas.sel(time=[5,6,7]).mean(dim='time').values  #mean over June, July, August
    delta_time_mean_upscaled = upscale(coarseness,[delta_time_mean])
    delta_time_mean = expand_dims(delta_time_mean_upscaled[0],axis=0)
    if ff == 0:
        delta_ens[var] = delta_time_mean
    else:
        delta_ens[var] = append(delta_ens[var],delta_time_mean,axis=0)
```

#### Reshape GSA anomaly

```python
gsa_anomaly = (SENSD_spatial[var]['gsa_range'] 
               - SENSD_spatial[var]['gsa_range'].mean(axis=0)
              ).reshape(append(-1,wrflon_upscaled.shape))
```

### Determine the warming reduction factor that allows us to say that warming is greater than uncertainy via the Mann-Whitney test

```python
def qt_helper(qtile): # get quantiles evenly spaced from 0.05 to 0.95 with length of proj ensemble
    return quantile(gsa_anomaly,qtile,axis=0)

pool = Pool(processes=delta_ens[var].shape[0])
qtiles = linspace(0.05,0.95,delta_ens[var].shape[0])
gsa_qtls = pool.map(qt_helper,qtiles)
```

```python
# put into numpy array
gsa_qtls_np = expand_dims(gsa_qtls[0],axis=0)
for gsa_qtl in gsa_qtls[1::]:
    gsa_qtls_np = append(gsa_qtls_np,expand_dims(gsa_qtl,axis=0),axis=0)
```

```python
# perform the mann-whitney U test
def mw_helper(reduce_fac): # the func called in child processes
    stat, pvalue = mannwhitneyu(
                reduce_fac*delta_ens[var],gsa_qtls_np,alternative='greater',axis=0,
            )
    
    #stat = zeros((gsa_anomaly.shape[1::]))
    #pvalue = zeros((gsa_anomaly.shape[1::]))
    #for xi in range(gsa_anomaly.shape[1]):
    #    for yi in range(gsa_anomaly.shape[2]):
    ##        stat[xi,yi], pvalue[xi,yi] = ks_2samp(
    #        stat[xi,yi], pvalue[xi,yi] = brunnermunzel(
    #            reduce_fac*delta_ens[var][:,xi,yi],gsa_anomaly[0::10,xi,yi],alternative='greater',
    #        )
    #print(stat[pvalue <= 0.05].mean())
    stat[pvalue >= 0.05] = 0
    return stat

pool = Pool(processes=20)
reduce_facs = linspace(0.01,0.20,20)
stat_results = pool.map(mw_helper, reduce_facs)
```

```python
warming_needed = 0*delta_ens[var].mean(axis=0)
for rr, res in enumerate(stat_results):
    warming_needed_temp = reduce_facs[rr]*delta_ens[var].mean(axis=0)
    warming_needed[res < +1e-6] = warming_needed_temp[res < +1e-6]
    #warming_needed[res > -1e-6] = warming_needed_temp[res > -1e-6]
    
# check that no values are higher
print((res < +1e-6).sum())
#print((res > -1e-6).sum())
```
